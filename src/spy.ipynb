```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import os
```

```python
start = "2023-01-01"
end = "2024-06-30"

output_folder = "../input/"
```

```python
indices = {
    "000001_SS": "000001.SS",
    "399001_SZ": "399001.SZ",
    "HSI": "^HSI",
    "IXIC": "^IXIC",
    "FTSE": "^FTSE",
    "GDAXI": "^GDAXI",
    "FCHI": "^FCHI",
    "600256_SS": "600256.SS"

}
```

```python
data = {}
```

```python
for index, ticker in indices.items():
    data[index] = yf.download(ticker, start=start, end=end)
    print(f"已获取 {index} 的数据")

for index, df in data.items():
    df.to_csv(os.path.join(output_folder, f"{index}.csv"))
    print(f"已保存 {index} 的数据到CSV文件")
```

```python

indices = ["000001_SS", "399001_SZ", "HSI", "IXIC", "FTSE", "GDAXI", "FCHI", "600256_SS"]


plt.figure(figsize=(14, 8))

# 循环加载
for index in indices:
    df = pd.read_csv(os.path.join(output_folder, f"{index}.csv"), index_col='Date', parse_dates=True)
    plt.plot(df.index, df['Close'], label=index)
plt.title('Stock_Trends')
plt.xlabel('DATE')
plt.ylabel('close')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 假设文件名为 'data.csv'，日期字段为 'date'
filename = 'SC0.csv'
date_column = 'date'

# 读取CSV文件
df = pd.read_csv(filename)

# 将日期字段转换为日期格式
df[date_column] = pd.to_datetime(df[date_column])

# 定义日期范围
start_date = '2023-01-01'
end_date = '2024-06-30'

# 筛选日期范围内的数据
filtered_df = df[(df[date_column] >= start_date) & (df[date_column] <= end_date)]

# 将日期字段转换回原始格式（假设原始格式为 'YYYY/M/D'）
filtered_df[date_column] = filtered_df[date_column].dt.strftime('%Y/%-m/%-d')

# 保存筛选后的数据到新的CSV文件
output_filename = 'filtered_data_SC0.csv'
filtered_df.to_csv(output_filename, index=False)

print(f"数据已保存到 {output_filename}")
```
